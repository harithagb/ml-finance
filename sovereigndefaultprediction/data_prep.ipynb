```python
import difflib
from dataclasses import dataclass
```

```python
@dataclass()
class FuzzyMerge:
    """
        Works like pandas merge except also merges on approximate matches.
    """
    left: pd.DataFrame
    right: pd.DataFrame
    left_on: str
    right_on: str
    how: str = "inner"
    cutoff: float = 0.7

    def main(self) -> pd.DataFrame:
        temp = self.right.copy()
        temp[self.left_on] = [
            self.get_closest_match(x, self.left[self.left_on]) for x in temp[self.right_on]
        ]

        return self.left.merge(temp, on=self.left_on, how=self.how)

    def get_closest_match(self, left: pd.Series, right: pd.Series) -> str or None:
        matches = difflib.get_close_matches(left, right, cutoff=self.cutoff)

        return matches[0] if matches else None
```

```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
merged = FuzzyMerge(left=df_wdi_fmt[['Country Name']].rename({'Country Name':'Country Name (WDI)'}, axis = 1).drop_duplicates(), 
                    right=df_weo[['Country']].rename({'Country':'Country Name (WEO)'}, axis=1).drop_duplicates(), 
                    left_on="Country Name (WDI)", right_on="Country Name (WEO)", how = 'inner',cutoff = 0.5).main()

sim_score = []

for idx,row in merged.iterrows():
    sim_score.append(similar(row['Country Name (WDI)'], row['Country Name (WEO)']))

merged['sim_score'] = sim_score
merged.sort_values(by =['Country Name (WDI)','sim_score'], ascending = [True,False], inplace = True)
```

```python
merged_countries = merged.groupby(by = 'Country Name (WDI)', as_index = False).agg({'Country Name (WEO)':lambda x: x.head(1)})
```

```python
merged_countries.head()
```

```python
merged_countries = FuzzyMerge(left= merged_countries, right = df_gfd_fmt[['Country Name']].rename({'Country Name':'Country Name (GFD)'}, axis = 1).drop_duplicates(), 
                    left_on="Country Name (WDI)", right_on="Country Name (GFD)", how = 'left',cutoff = 0.7).main()

merged_countries
```

```python
merged_countries = FuzzyMerge(left= merged_countries, right = df_imf_yrly_fmt[['Country Name']].rename({'Country Name':'Country Name (IMF)'}, axis = 1).drop_duplicates(), 
                    left_on="Country Name (WDI)", right_on="Country Name (IMF)", how = 'left',cutoff = 0.6).main()
```
